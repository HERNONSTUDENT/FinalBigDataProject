```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, trim, lower
import re

# Initialize Spark Session
spark = SparkSession.builder \
    .appName("Bronze to Silver Data Cleansing") \
    .getOrCreate()


#bronze_path = "Tables/airlines_raw"
#df_raw = spark.read.format("delta").load(bronze_path)

df_raw = spark.read.table("Bronze_Weather_Raw.airlines_raw")


print(df_raw)
df_raw.show()
df_raw.printSchema()
df_raw.dropna() # Drop rows with missing values


df_raw.write.format("delta").mode("overwrite").saveAsTable("airlines_refined")
```
